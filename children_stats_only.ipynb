```python
occupation = pd.DataFrame(df['Occupation'].value_counts())
occupation = occupation.reset_index()
occupation = occupation.rename(columns = {'index':'occupation', 'Occupation':'count'})

for oc in occupation:
    occupation['perc'] = (occupation['count'] / df.shape[0] )*100
```

```python
occupation
```

```python
plt.figure(figsize = (12,6))
plt.style.use('default')

cmap=plt.get_cmap('tab20c')
clr=[cmap(i) for i in np.linspace(0,1,5)]
explode = [0.1,0,0,0,0]

sns.despine()

plt.subplot(122)
pie = plt.pie(x = 'perc' , 
              labels = 'occupation' ,
              colors = clr,
              autopct = '%1.2f%%', 
              explode = explode,
              data = occupation,
              shadow=True)
```

```python
num_cols = [col for col in df.columns if
           df[col].dtype == 'int' or df[col].dtype == 'float']
num_cols
```

```python
plt.figure(figsize = (11,8))

for i,col in enumerate(num_cols):
    plt.subplot(5,5,i+1)
    sns.histplot( x = col, data = df, 
                 bins = 30, color = 'steelblue')
    plt.tight_layout()
```

```python
for col in num_cols:
    print(f'skewness for {col} :',df[col].skew())
```

```python
plt.figure(figsize = (11,10))
plt.style.use('seaborn')
for i,col in enumerate(num_cols):
    plt.subplot(5,5,i+1)
    sns.boxplot( y = col, data = df, 
                color = 'steelblue',
                showmeans = True)
    plt.tight_layout()
```
