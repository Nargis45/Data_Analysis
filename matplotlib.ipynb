```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y,c='g')
plt.xlabel('x-axis')
plt.ylabel('y-label')
plt.title('graph in 2D')
plt.savefig('Test.png')
```

```python
y=x*x
```

```python
plt.plot(x,y,'r*--',linestyle='dashed',linewidth=2,markersize=12)
plt.xlabel('x-axis')
plt.ylabel('y-label')
plt.title('graph in 2D')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'bo-')
plt.subplot(2,2,3)
plt.plot(x,y,'g*')
plt.subplot(2,2,4)
plt.plot(x,y,'--g*')
```

```python
x=np.arange(1,11)
y=3*x+5
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-label')
plt.title('graph in 2D')
```

```python
np.pi
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-label')
plt.title('graph in 2D')
plt.show()
```

```python
x=np.arange(0,5*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)
plt.subplot(2,1,1)
plt.plot(x,y_sin)
plt.title('sine')

plt.subplot(2,1,2)
plt.plot(x,y_cos)
plt.title('cosine')
```

```python
x=[2,8,10]
y=[11,16,9]
x2=[3,9,11]
y2=[16,15,7]
plt.bar(x,y,align='center')
plt.bar(x2,y2,color='green')
plt.xlabel('x-axis')
plt.ylabel('y-label')
plt.title('Bar graph')
```

```python
a=np.array([22,12,15,19,33,35,66,78,6,56,25,29,45,98])
plt.hist(a)
plt.title('histogram')
```

```python
data=[np.random.normal(0, std, 100) for std in range(1, 4)]
plt.boxplot(data, vert=True, patch_artist=True)
```

```python
labels='python', 'c++', 'Ruby', 'Java'
sizes=[215, 130, 245, 210]
colors='gold','yellowgreen','lightcoral','lightskyblue'
explode=(0.1,0,0,0)
plt.pie(sizes,explode=explode, labels=labels, colors=colors,autopct='%1.1f%%', shadow=True)
plt.axis('equal')
```
